```python
from pycaret.classification import *
import pandas as pd

url = 'https://raw.githubusercontent.com/coen2812023/Classification/main/customer_churn.csv'
df = pd.read_csv(url, index_col=0)

s = setup(df, target = 'churned')
```

```python
best = compare_models()
```

```python
best
```

```python
evaluate_model(best)
```
